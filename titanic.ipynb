```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os

#gender_sub = pd.read_csv("../input/titanic/gender_submission.csv")
test = pd.read_csv("../input/titanic/test.csv")
train = pd.read_csv("../input/titanic/train.csv")

N = 891 # data number

X = np.ones((N,3))
Y = train['Survived']

# x1: Pclass, x2: Sex(male==1, female==0)
for i in range(N):
    X[i,1] = train['Pclass'][i]
    if train['Sex'][i] == "male":
        X[i,2] = 1
    else:
        X[i,2] = 0

# Estimation
Beta = np.linalg.inv(X.T @ X) @ X.T @ Y

Ys = np.array(Y)

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter3D(X[:,1], X[:,2], Ys)

Xd, Yd = np.meshgrid(np.arange(1, 3, 0.01), np.arange(0,1,0.01))
Zd = Beta[1] * Xd + Beta[2] * Yd + Beta[0]
ax.plot_surface(Xd, Yd, Zd, alpha = 0.5, color = "red") #alphaで透明度を指定
ax.set_xlabel("Pclass")
ax.set_ylabel("Sex")
ax.set_zlabel("Survived")

plt.show()

TestN = 418
Y_Test = np.zeros(TestN,dtype=int)

for i in range(TestN):
    sex = 0
    if test['Sex'][i] == "male":
        sex = 1
    ytmp = Beta[0] + Beta[1] * test['Pclass'][i] + Beta[2] * sex
    
    if ytmp < 0.5:
        Y_Test[i] = 0
    else:
        Y_Test[i] = 1

# PassengerIdを取得
PassengerId = np.array(test["PassengerId"]).astype(int)
# Y_Test(予測データ）とPassengerIdをデータフレームへ落とし込む
my_solution = pd.DataFrame(Y_Test, PassengerId, columns = ["Survived"])
# my_tree_one.csvとして書き出し
my_solution.to_csv("mySoluve.csv", index_label = ["PassengerId"])

Y_Test

# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import os

def Lacking_table(df): 
        null_val = df.isnull().sum()
        percent = 100 * df.isnull().sum()/len(df)
        kesson_table = pd.concat([null_val, percent], axis=1)
        kesson_table_ren_columns = kesson_table.rename(
        columns = {0 : '欠損数', 1 : '%'})
        return kesson_table_ren_columns

test = pd.read_csv("../input/titanic/test.csv")
train = pd.read_csv("../input/titanic/train.csv")


train["Age"] = train["Age"].fillna(train["Age"].median())
train["Embarked"] = train["Embarked"].fillna("S")

train["Sex"][train["Sex"] == "male"] = 0
train["Sex"][train["Sex"] == "female"] = 1
train["Embarked"][train["Embarked"] == "S" ] = 0
train["Embarked"][train["Embarked"] == "C" ] = 1
train["Embarked"][train["Embarked"] == "Q"] = 2

Lacking_table(train)

N = 891
Relative_Data_train = np.zeros(N,dtype=int)

for i in range(N):
    Relative_Data_train[i] = train['SibSp'][i] + train['Parch'][i]

X = np.ones((N,6))
Y = train['Survived']

for i in range(N):
    X[i,1] = train['Pclass'][i]
    X[i,2] = train['Sex'][i]
    X[i,3] = train['Age'][i]
    X[i,4] = train['Fare'][i]
    X[i,5] = Relative_Data_train[i]
    
# Estimation
Beta = np.linalg.inv(X.T @ X) @ X.T @ Y

test["Age"] = test["Age"].fillna(train["Age"].median())
test["Embarked"] = test["Embarked"].fillna("S")

test["Sex"][test["Sex"] == "male"] = 0
test["Sex"][test["Sex"] == "female"] = 1
test["Embarked"][test["Embarked"] == "S" ] = 0
test["Embarked"][test["Embarked"] == "C" ] = 1
test["Embarked"][test["Embarked"] == "Q"] = 2

TestN = 418
Y_Test = np.zeros(TestN,dtype=int)

Relative_Data_test = np.zeros(TestN,dtype=int)

for i in range(TestN):
    Relative_Data_test[i] = test['SibSp'][i] + test['Parch'][i]

for i in range(TestN):
    ytmp = Beta[0] + Beta[1] * test['Pclass'][i] + Beta[2] * test['Sex'][i] + Beta[3] * test['Age'][i] + Beta[4] * test['Fare'][i] + Beta[5] * Relative_Data_test[i]
    if ytmp < 0.5:
        Y_Test[i] = 0
    else:
        Y_Test[i] = 1

# PassengerIdを取得
PassengerId = np.array(test["PassengerId"]).astype(int)
# Y_Test(予測データ）とPassengerIdをデータフレームへ落とし込む
my_solution = pd.DataFrame(Y_Test, PassengerId, columns = ["Survived"])
# my_tree_one.csvとして書き出し
my_solution.to_csv("mySoluve3.csv", index_label = ["PassengerId"])
```
